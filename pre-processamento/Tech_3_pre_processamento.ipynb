```python
from datasets import load_dataset

ds = load_dataset("LuizfvFonseca/trn_limpo_parte_1_de_70")

print(ds)

train_ds = ds['train']

df = train_ds.to_pandas()
df.head()
```

```python
# Exibe as colunas
print("Colunas no DataFrame:")
print(df.columns)

# Exibe o tamanho do dataframe (número de linhas e colunas)
print("\nTamanho do DataFrame:")
print(f"Número de linhas: {df.shape[0]}")
print(f"Número de colunas: {df.shape[1]}")
```

```python
# Remove as colunas 'target_ind' and 'target_rel'
df = df.drop(columns=['target_ind', 'target_rel'])

df.head()
```

```python
# Remove rows with any null values
df = df.dropna()
```

```python
# Exibe as colunas
print("Colunas no DataFrame:")
print(df.columns)

# Exibe o tamanho do dataframe (número de linhas e colunas)
print("\nTamanho do DataFrame:")
print(f"Número de linhas: {df.shape[0]}")
print(f"Número de colunas: {df.shape[1]}")
```

```python
# Remove linhas com valores em brancos
df = df[~df.map(lambda x: isinstance(x, str) and x.strip() == '').any(axis=1)]

# Exibe o tamanho do DataFrame após o cleanup
print("\nTamanho do DataFrame após o cleanup:")
print(f"Número de linhas: {df.shape[0]}")

# Exibe os 5 primeiros registros
df.head()
```
